```python
for f in tqdm(files):
  img = cv2.imread(f)
  if img.shape[0] > img_size or img.shape[1] > img_size:
    img = cv2.resize(img , (img_size,img_size))
  # padd = np.ones((64,64,3), dtype='uint8')*255
  padd = np.zeros((img_size,img_size,3), dtype='uint8')
  # plt.imshow(img)
  # plt.show()
  w,h = img.shape[:2]
  # print(w,h)
  padd[ (img_size-w)//2:(img_size-w)//2 + w , (img_size-h)//2:(img_size-h)//2 + h,:] = img
  # plt.imshow(padd)
  # plt.show()
  name = os.path.basename(f)
  name = name.split('.')[0]
  # print(f'{dest}/{name}.png')
  cv2.imwrite(f'{dest}/{name}.png', padd)
```
