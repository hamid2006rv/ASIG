```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = tf.keras.models.load_model('/content/drive/MyDrive/CancerCell/24 April/NewDataSet_Ravaee_Feb2024/Dataset1 Scripts/result 5.13.24/GE W-cGAN Generator v2.h5')
```

```python
n = 300
latent_dim = 128
num_features = 1
```

```python
# output_dir = '/content/drive/MyDrive/CancerCell/24 April/NewDataSet_Ravaee_Feb2024/Dataset1 Scripts/result 5.13.24/output4'
# !mkdir "$output_dir"
```

```python
output_dir = './output4'
```

```python
classes = [0,1,2]
classes_normal = np.asarray(classes) / 3 - 0.5
```

```python
classes_normal
```

```python
for j in range(len(classes)):
  features = [classes_normal[j]] * n
  _noise = tf.random.normal(shape=(len(features), latent_dim))
  labels = np.array(features).reshape((len(features),num_features))

  # _noise_and_labels = tf.concat([_noise, labels], 1)
  fake_images = G.predict([_noise, labels])

  fake_images = fake_images - 0.1 ######

  fake_images = np.clip(fake_images, -1 ,1)
  fake_images = fake_images * 0.5 + 0.5
  fake_images *= 255.0
  converted_images = fake_images.astype(np.uint8)
  for i in range(len(features)):
    # plt.imshow(converted_images[i])
    # plt.show()
    # print(output_dir + f'/{i}_class_{j}.png')
    plt.imsave(output_dir + f'/{i}_class_{j}.png',converted_images[i])
```
